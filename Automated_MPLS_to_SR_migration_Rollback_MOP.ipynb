#### Remove SR from Core and PEs

```python
!ansible-playbook roles/SR_rollback/Remove_SR/tasks/SR-Config-Rollback.yml --extra-vars "target=core,PE"

!ansible-playbook roles/Show_commands/tasks/XR-show-cmd.yml --extra-vars "target=core,PE commands=['sh running-config router isis xr address-family ipv4 unicast','sh running-config router isis xr interface loopback 0']"



!ansible-playbook roles/Show_commands/tasks/ntc-show-cmd.yml --extra-vars "target=PE,core commands='sh mpls forwarding'"


import ast
import pandas as pd

def highlight_cols(s):
    color = 'lightyellow'
    return 'background-color: %s' % color
hosts = !ansible PE,core --list-hosts
data = [line.rstrip("\n\r") for line in open("roles/Show_commands/output/show_cmd_ouput_json.txt", "r")]
data = data[0]
data = ast.literal_eval(data)
total_host = len(data)
for x in range(total_host):
    data0 = data[x]
    locals()['df{0}'.format(x)] = pd.DataFrame(data0)
    locals()['df{0}'.format(x)] = locals()['df{0}'.format(x)][["prefix","local_label","out_label","next_hop","out_intf","bytes_switched"]]
    locals()['df{0}'.format(x)] = locals()['df{0}'.format(x)].style.set_caption('Host: '+ str(hosts[x+1])).applymap(highlight_cols, subset=pd.IndexSlice[:, ['prefix', 'bytes_switched']])
    display(locals()['df{0}'.format(x)])


print "="*30
print "Done. Rollback Complete"
print "="*30
```
